```python
# Importing Necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Uploading the data
data = pd.read_csv('titanic.csv')
df = pd.DataFrame(data)
```

```python
# Viewing the dataset
df.head()
```

```python
# Performing EDA
```

```python
# Summerizing the dataset
df.shape
```

```python
df.describe()
```

```python
df.dtypes
```

```python
# Handling Missing Values
```

```python
df.isnull().sum()
```

```python
# Age Column
df['Age'].mean()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
# Embarked column
df['Embarked'].mode()[0]
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# encoding categorical data
# Sex column
```

```python
df['Sex'] = df['Sex'].map({'male':0,'female':1})
```

```python
df.head()
```

```python
# Scaling Numerical Features
```

```python
# Fare Column
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df['Fare'] = scaler.fit_transform(df[['Fare']])
```

```python
df
```

```python
# Removing duplicates
```

```python
df = df.drop_duplicates()
```

```python
# Text Data Preprocessing
```

```python
df['Name'] = df['Name'].str.lower()
```

```python
df.head()
```

```python
# Sample Visualization
```

```python
# How does gender impact survival rate
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Sex',y='Survived',data=df)
plt.show()
```

```python
# Impact of age on survival
```

```python
sns.histplot(data=df, x='Age', hue='Survived')  # Use hue to differentiate by 'Survived'
plt.title('Age vs Survival')
plt.show()
```
